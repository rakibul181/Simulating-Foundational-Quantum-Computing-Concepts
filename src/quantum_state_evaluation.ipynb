# Quantum State Evolution: Simulation and Visualization

## 1. Introduction

This section presents the simulation and analysis of quantum state evolution for a single-qubit quantum circuit using Qiskit. The circuit consists of a Hadamard gate followed by a Z gate applied to an initial \(|0\rangle\) state.

## 2. Quantum Circuit and State Evolution

The quantum circuit applies a Hadamard gate ($H$) to the qubit, creating a superposition state, followed by a Pauli-Z ($Z$) gate which introduces a phase shift to the $|1\rangle$ component:

\[
|0\rangle \xrightarrow{H} \frac{1}{\sqrt{2}}\left(|0\rangle + |1\rangle\right) \xrightarrow{Z} \frac{1}{\sqrt{2}}\left(|0\rangle - |1\rangle\right)
\]

The final state, $|-\rangle$, is a balanced superposition with a relative phase difference.

## 3. Statevector Output

The Qiskit simulation returns the final quantum state as:

\[
|\psi\rangle = 0.7071|0\rangle - 0.7071|1\rangle = \frac{1}{\sqrt{2}}\left(|0\rangle - |1\rangle\right)
\]

This state has real-valued amplitudes, reflecting the effect of the applied quantum gates.


## 4. Visualization and Interpretation

### a) Bloch Sphere Representation

*The Bloch sphere plot shows the quantum state vector located on the equator along the negative x-axis, confirming the \(|-\rangle\) state. This geometric visualization demonstrates the phase flip introduced by the Z gate.*

### b) State City Visualization

*The State City visualization decomposes the quantum state into real and imaginary parts:*
- *The real amplitude plot displays two bars, one positive (\(|0\rangle\)) and one negative (\(|1\rangle\)), matching the expected statevector.*
- *The imaginary amplitude plot shows all components at zero, indicating a purely real state.*

## 5. Physical Meaning

Although the measurement probabilities in the computational basis are balanced (50% for \(|0\rangle\) and 50% for \(|1\rangle\)), the phase difference in the superposition is crucial for further quantum computations and interference phenomena.

## 6. Conclusion

This simulation and visualization demonstrate the effects of quantum gates on single-qubit states. The combination of Bloch sphere and State City plots provides a clear understanding of quantum phase and superposition, highlighting essential concepts for quantum computing.


```python
%matplotlib inline
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_city
import matplotlib.pyplot as plt

# Create a quantum circuit with 2 qubits
qc = QuantumCircuit(2)

# Apply Hadamard gate to the first qubit
qc.h(0)

# Apply CNOT gate (control=0, target=1)
qc.cx(0, 1)

# Simulate the quantum circuit statevector
statevector = Statevector.from_instruction(qc)

# Print the statevector
print("Bell State vector:", statevector)

# Visualize the quantum state using Bloch sphere
plot_bloch_multivector(statevector)

# Visualize the state using state city representation
plot_state_city(statevector)

# Show the plots
plt.show()
```

## For 3 qubit

```python
# Import libraries
%matplotlib inline
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_city
import matplotlib.pyplot as plt

# 1. Create a quantum circuit with 3 qubits
qc3 = QuantumCircuit(3)
for q in range(3):
    qc3.h(q)  # Hadamard on each qubit
    qc3.z(q)  # Z gate on each qubit

qc3.draw('mpl')  # Optional: visualize circuit

# 2. Get the final statevector
statevector3 = Statevector.from_instruction(qc3)
print("Final state vector (3 qubits):", statevector3)


# 3. State city plot
plot_state_city(statevector3)
plt.title("State City Visualization (3 Qubits)")
plt.show()
```
