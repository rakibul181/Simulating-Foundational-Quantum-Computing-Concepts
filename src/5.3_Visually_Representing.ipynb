## Single-Qubit Gates Visualizations
To see the each of the gates, we will take a single circuit with apply a different gate to each of the qubits to plot the values of each of those qubits on the blocsphere.
### X Gate

```python
import numpy as np

# Import Qiskit core
from qiskit import QuantumCircuit

# For visualization
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt
# Import Aer from qiskit-aer
from qiskit_aer import Aer
from qiskit_aer import AerSimulator
sim = AerSimulator()
backend = Aer.get_backend('statevector_simulator')
```

```python
# Get the statevector from the circuit
from qiskit.quantum_info import Statevector
```

```python
# Create a single-qubit circuit and apply an X-gate to get |1‚ü©
qc = QuantumCircuit(1)

#before apply sate
state1 = Statevector.from_instruction(qc)
print(state1)

qc.x(0)

# after apply Gate state
state2 = Statevector.from_instruction(qc)
print(state2)
qc.draw("mpl")
```

```python
 # Plot the Bloch sphere
plot_bloch_multivector(state1, title="Bloch Sphere of |0‚ü© State")
```

```python
plot_bloch_multivector(state2, title="Bloch Sphere of |1‚ü© State")
```

```python
qc.measure_all()
qc.draw('mpl')
```

```python
job = sim.run(qc)
counts = job.result().get_counts()
print(counts)
display(plot_histogram(counts))
    
```


### The Y gate

```python
# let's do an X-gate on a |0> qubit
qc = QuantumCircuit(1)
state1 = Statevector.from_instruction(qc)

print("Initail", state1)
# Let's see the result

qc.y(0)                 # apply the y gate

state2  = Statevector.from_instruction(qc)
print('Final',state2)

qc.draw('mpl')
```

```python

plot_bloch_multivector(state1, "initail state")
```

```python
plot_bloch_multivector(state2,"Final state")
```

```python
qc.measure_all()
qc.draw('mpl')
```

```python
job = sim.run(qc)
counts = job.result().get_counts()
print('Probability',counts)
display(plot_histogram(counts))
```
